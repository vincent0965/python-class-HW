numpy array 用法

```python
import numpy as np

array1 = np.arange(6).reshape(2, 3)
array2 = np.arange(1, 4)
print(array1)
print()
print(array2)
```

```python
# array運算
print(array1 + array2)
print(np.add(array1, array2))
print()
print(array1 - array2)
print(np.subtract(array1, array2))
print()
print(array1 * array2)
print(np.multiply(array1, array2))
print()
print(array1 / array2)
print(np.divide(array1, array2))
print()
print(array1 % array2)
print(np.mod(array1, array2))
print()
print(array1 ** array2)
print(np.power(array1, array2))
print()
```

```python
array1 = np.array([0.3, 0.5, 0.75, 1, 100])
print(np.reciprocal(array1))  # 求倒數
array2 = np.array([1., 2, 3, 4])  # 其中包含 float 因此只操作 float
print(np.reciprocal(array2))
```

```python
print(round(19.569))  # 無條件進位
print(round(19.569, 1))  # 無條件進位到小數點第一位
```

```python
# 用法同 round、floor、ceil
array1 = np.linspace(13, 15, 6)
print(array1)
print(np.around(array1, 1))
print(np.around(array1, -1))
print()
print(np.floor(array1))
print()
print(np.ceil(array1))
```

```python
array1 = np.arange(6).reshape(2, 3)
print(array1)
print()
print(np.sum(array1))
print(np.sum(array1, axis = 0)) # 將每個row總和 (相同位置的row:0、3；1、4；2、5)
print(np.sum(array1, axis = 1))  # 將每個columu總和(相同位置的columu:0、1、2；3、4、5)
print()
print(np.max(array1))  # 選出array中最大的
print(np.max(array1, axis = 0)) # 選出row中最大的(相同位置的row:0、3；1、4；2、5)
print(np.max(array1, axis = 1))  # 選出columu中最大的(相同位置的columu:0、1、2；3、4、5)
print()
print(np.min(array1))  # 選出array中最小的
print(np.min(array1, axis = 0)) # 選出row中最小的(相同位置的row:0、3；1、4；2、5)
print(np.min(array1, axis = 1))  # 選出columu中最小的(相同位置的columu:0、1、2；3、4、5)
```

```python
np.random.seed(1)
array1 = np.random.randint(1, 11, (2, 3))
print(array1)
print()
print(np.mean(array1))
#平均
print(np.mean(array1, axis = 0))
print(np.mean(array1, axis = 1))
print()
print(np.median(array1))
#中位數
print(np.median(array1, axis = 0))
print(np.median(array1, axis = 1))
```

```python
array1 = np.arange(1, 5)
print(array1)
print(np.mean(array1))
print(np.average(array1))
# 每個數字的權重
print(np.average(array1, weights = [2, 2, 1, 1]))
# 回傳分母 (2+2+1+1)
print(np.average(array1, weights = [2, 2, 1, 1], returned = True))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline  
# 僅創建在jupyter的環境底下 jupyter magic command 畫圖模組

# 圖形大小 後面可以給予圖形控制(ex:顏色)
plt.figure(figsize = (8, 6), facecolor = "pink")
# 四個點都會標在圖形上 plot為折線圖 前為x 後為y 可以加入控制方法(顏色 點 線條等)
plt.plot([2, 4, 6, 8], [1, 5, 2, 6], "bo--")
plt.show()
```

```python
# 圖形設計(折線圖)
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(10)
np.random.seed(1234)
y1 = np.random.randint(0, 10, 10)
y2 = np.random.randint(0, 10, 10)
plt.figure(figsize = (12, 8))
# 可以同誰放上兩條線 並給予線條樣式修正
plt.plot(x, y1,"go-", label = "Green")
plt.plot(x, y2, "m^--", label = "Score")
# 顏色可以加上色碼調整 並加入各種可調整的內容 線長 線寬 點 顏色 透明度
plt.plot([1, 2, 3], [4, 5, 6], 
         color = (50/255, 10/255, 150/255), 
         linestyle = ":", linewidth = 4, 
         marker = "*", markersize = 15,
        label = "Ex1")
# 放入字串
plt.title("This is a plot sample.")
plt.xlabel("X axis")
plt.ylabel("Y axis")
# 座標邊界
plt.xlim(-1, 10)
plt.ylim(-2, 10)
# 網格線
plt.grid()
# 放置兩邊的標記
plt.xticks([1, 5, 9])
plt.yticks([2, 5, 8])
# 圖標
plt.legend()
plt.show()





```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# 長條圖 內有許多的控制變數 如:項目 值 圖形寬度 顏色 外框(?)
plt.bar(range(4), [12, 34, 13, 3], 0.5, color = "navy", edgecolor = "red")
plt.show()
```

```python
# 堆疊長條圖
# np.random.seed(12)
men = np.random.randint(15, 35, 6)
women = np.random.randint(15, 35, 6)
# 以shape來顯示 x產生的範圍
x = np.arange(men.shape[0])

print(men)  # men的array
print(men.shape)  # men的型態(taple)
print(men.shape[0])  # 用taple的形式來決定我的x產生方式

# x,y 的值以及圖示
plt.bar(x, men, label = "men")
# bottom 在底部留空位 圖才不會重疊
plt.bar(x, women, bottom = men, label = "women")
# y軸拉高
plt.ylim(0, 80)
# x的名稱
plt.xticks(x, ("G1", "G2", "G3", "G4", "G5", "G6"))
plt.legend()
plt.show()
```

```python
# 堆疊長條圖
# np.random.seed(12)
men = np.random.randint(15, 35, 6)
women = np.random.randint(15, 35, 6)
# 以shape來顯示 x產生的範圍
x = np.arange(men.shape[0])
# 改變我的寬度
width = 0.3

# x,y 的值以及圖示
plt.bar(x - width/2, men, width = width, label = "men")
# bottom 在底部留空位 圖才不會重疊
plt.bar(x + width/2, women, width = width, label = "women")
# 加入width 可以控制線寬 讓圖變成分組並行且可以重疊

# y軸拉高
plt.ylim(0, 40)
# x的名稱
plt.xticks(x, ("G1", "G2", "G3", "G4", "G5", "G6"))
plt.legend()
plt.show()
```

```python
# 散布圖
# 前後互相對應產生點 亦可以改變每個點的圖形值 alpha表示透明度
plt.scatter([1, 1, 2, 2], [1, 2, 3, 4], 
            s = [100, 200, 300, 500], c = ["r", "b", "g", "m"],
           alpha = 0.5)
plt.show()
```

```python
# np.random.seed(1)
n = 200
x = np.random.rand(n)
y = np.random.rand(n)

# 產生不同顏色
color = np.random.rand(n)
# 產生不同的點大小
size = (np.random.rand(n) * 10) ** 3

# 每個點可以設定不同顏色以及大小等
plt.scatter(x, y , c = color, s = size, alpha = 0.5)
plt.show()
```
